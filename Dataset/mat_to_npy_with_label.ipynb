```python
import numpy as np
import os
import pandas as pd
import scipy.io
```

```python
label_tag = { 'N': 0, 'A': 1, 'O': 2, '~': 3}

# mat 파일로 저장된 데이터 폴더 경로
data_path = 'training2017_9s'
# REFERENCE-original.csv 파일의 경로
reference_label = 'training2017/REFERENCE-original.csv'
# 데이터 저장 위치 및 이름
save_data = 'training_9s'
# label 저장 위치 및 이름
save_label = 'training_9s_labels'

# data list
data_file = []
# label list
labeling = []

# 데이터의 경로에서 저장되어있는 데이터의 이름을 수집
data_list = os.listdir(data_path)
# 파일 번호가 섞여있을 것을 대비해 sorted로 정렬
data_list_sort = sorted(data_list)

# REFERENCE label 파일을 이용하여 labeling
labeling_name = pd.read_csv(reference_label, names=('name', 'label'))


# REFERENCE의 name과 파일이름이 같은지 확인
# 확인한 후 Noise인지 확인하고 아니면 data_file에 저장 및 labeling에 label 저장
count = 0
for i in data_list_sort[:-1]:
    if labeling_name['name'][count] in i:
        if labeling_name['label'][count] != '~':
            labeling.append(label_tag[labeling_name['label'][count]])
            mat_file = scipy.io.loadmat(data_path + '/' + i)
            data_file.append(mat_file['u'])
        else:
            continue
    else:
        count+=1
        if labeling_name['label'][count] != '~':
            labeling.append(label_tag[labeling_name['label'][count]])
            mat_file = scipy.io.loadmat(data_path + '/' + i)
            data_file.append(mat_file['u'])
        else:
            continue
    print(i)

np_data_file = np.array(data_file)
np_labeling = np.array(labeling)

np.save(save_data, np_data_file)
np.save(save_label, np_labeling)

print('저장완료')
```
